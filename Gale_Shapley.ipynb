```python
list_preferencia_hospitales = {
'Atlanta' : ['Wayne', 'Val', 'Yolanda', 'Zeus', 'Xavier'],
'Boston'  : ['Yolanda', 'Wayne', 'Val', 'Xavier', 'Zeus'],
'Chicago' : ['Wayne', 'Zeus', 'Xavier', 'Yolanda', 'Val'],
'Detroit' : ['Val', 'Yolanda', 'Xavier', 'Wayne', 'Zeus'],
'El Paso' : ['Wayne', 'Yolanda', 'Val', 'Zeus', 'Xavier'],
}
 
list_preferencia_estudiantes = {
'Val'     : ['El Paso', 'Atlanta', 'Boston', 'Detroit', 'Chicago'],
'Wayne'   : ['Chicago', 'Boston', 'Detroit', 'Atlanta', 'El Paso'],
'Xavier'  : ['Boston', 'Chicago', 'Detroit', 'El Paso', 'Atlanta'],
'Yolanda' : ['Atlanta', 'El Paso', 'Detroit', 'Chicago', 'Boston'],
'Zeus'    : ['Detroit', 'Boston', 'El Paso', 'Chicago', 'Atlanta'],
}

#Emparejamiento perfecto
conjunto_s = []
#lista de Hospitales (Hospitales sin Estudiantes)
hospitals = [hospital for hospital in list(list_preferencia_hospitales.keys())]

def stableMatch():
    while len(hospitals) > 0: #Validacion que ninguno hospital este sin un estudiante
        for hospital in hospitals:
            begin_matching(hospital)

def begin_matching(hospital):
    print('Hospital ' + hospital)
    for student in list_preferencia_hospitales[hospital]:
        taken_match = [pairing for pairing in conjunto_s if student in pairing]
        if not taken_match:
            conjunto_s.append([hospital, student])
            hospitals.remove(hospital)
            print(f"Posible pareja ({hospital}, {student})")
            break
        else:
            print(f"{student} ya tiene asignado el Hospital {hospital}")
            actual_match = list_preferencia_estudiantes[student].index(taken_match[0][0])
            posible_match = list_preferencia_estudiantes[student].index(hospital)
            if posible_match < actual_match:
                print(f"[Para {student}] {hospital} ocupa un lugar mas alto de preferencia que {taken_match[0][0]}")
                print(f"[Nueva pareja] Posible pareja ({hospital}, {student})")
                hospitals.remove(hospital)
                hospitals.append(taken_match[0][0])
                taken_match[0][0] = hospital
                break
            else:
                print(f"[Inestabilidad] {student} prefiere a {taken_match[0][0]}")

if __name__ == '__main__':
    stableMatch()
    print('Emparejamiento perfecto')
    print(conjunto_s)
```
